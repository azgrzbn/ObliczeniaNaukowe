```python
from sympy import *
import numpy as np

def calka_R(funkcja,a,b,N):
  podzial=np.linspace(a,b,N)
  podstawa=(b-a)/(N-1)
  wartosc=0
  for i in range(N-1):
    wartosc+=podstawa*abs(funkcja(podzial[i]))
  return wartosc

def sup(funkcja, a, b):
  moc_kand = [a,b]
  slab_kand = solve(diff(funkcja,x), x)
  for i in range (len(slab_kand)):
    if slab_kand[i]>a and slab_kand[i]<b :
      moc_kand.append(slab_kand[i])
    moc_kand[i]=funkcja.subs(x, moc_kand[i])
  return max(moc_kand)

def blad(funkcja, x, a, b):
  fun = lambdify(x, funkcja)
  M = sup(sin,a,b)
  N = floor((10**5)*M*((b-a)**2)/2)
  return (calka_R(fun, a, b, N))

x = symbols('x')
f = sin(x)
print(blad(f, x, 0, 1))
```
